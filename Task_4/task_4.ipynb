```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path_1 = ('D:\PROJECT\Webelight Interview Tasks\Task_4\images\images\cars.jpg')
image_path_2 = ('D:\PROJECT\Webelight Interview Tasks\Task_4\images\images\humans.jpg')
```

```python
# Function to display images
def display_image(image, title="Image"):
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

1. Image Loading

```python

image1 = cv.imread(image_path_1)
image2 = cv.imread(image_path_2)
```

2.Color Space Conversion:

```python
# Convert to grayscale
gray_image1 = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)
display_image(gray_image1, "Grayscale Image - Cars")

gray_image2 = cv.cvtColor(image2, cv.COLOR_BGR2GRAY)
display_image(gray_image2, "Grayscale Image - Humans")
```

```python
# Convert to HSV
hsv_image1 = cv.cvtColor(image1, cv.COLOR_BGR2HSV)
display_image(hsv_image1, "HSV Image - Cars")

hsv_image2 = cv.cvtColor(image2, cv.COLOR_BGR2HSV)
display_image(hsv_image2, "HSV Image - Humans")
```

3. Image Filtering

```python
# Apply Gaussian Blur
blur_image1 = cv.GaussianBlur(image1, (5, 5), 0)
display_image(blur_image1, "Gaussian Blur - Cars")

blur_image2 = cv.GaussianBlur(image2, (5, 5), 0)
display_image(blur_image2, "Gaussian Blur - Humans")
```

```python
# Apply custom kernel for image smoothing
kernel = np.ones((5, 5), np.float32) / 25
custom_smooth_image1 = cv.filter2D(image1, -1, kernel)
display_image(custom_smooth_image1, "Image Smoothing - Cars")

custom_smooth_image2 = cv.filter2D(image2, -1, kernel)
display_image(custom_smooth_image2, "Image Smoothing - Humans")
```

4.Edge Detection

```python
# Apply Canny edge detection
edges_image1 = cv.Canny(gray_image1, 100, 200)
display_image(edges_image1, "Canny Edges - Cars")

edges_image2 = cv.Canny(gray_image2, 100, 200)
display_image(edges_image2, "Canny Edges - Humans")
```

5.Contours

```python
# Find contours on the original image based on Canny edges
contours1, _ = cv.findContours(edges_image1, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
contoured_image1 = cv.drawContours(image1.copy(), contours1, -1, (0, 255, 0), 2)
display_image(contoured_image1, "Contours - Cars")

contours2, _ = cv.findContours(edges_image2, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
contoured_image2 = cv.drawContours(image2.copy(), contours2, -1, (0, 255, 0), 2)
display_image(contoured_image2, "Contours - Humans")
```

6. Histogram

```python
# Compute and display histogram of the grayscale image
plt.hist(gray_image1.ravel(), bins=256, color='gray', alpha=0.7)
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.title('Histogram - Cars')
plt.show()

plt.hist(gray_image2.ravel(), bins=256, color='gray', alpha=0.7)
plt.xlabel('Intensity Value')
plt.ylabel('Pixel Count')
plt.title('Histogram - Humans')
plt.show()
```

```python
# Load the pre-trained Haar Cascade face detector
face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Function to detect faces in an image
def detect_faces(image):
    # Convert the image to grayscale
    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    # Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    return image

# Load the image containing humans
image2 = cv.imread(image_path_2)

# Detect faces in the image of humans
faces_detected_image = detect_faces(image2)
display_image(faces_detected_image, "Faces Detected - Humans")
```

```python
cascade_src = 'cars.xml'

# Load the image
image_src = 'D:\PROJECT\Webelight Interview Tasks\Task_4\images\images\cars.jpg'
img = cv.imread(image_src)

# Create a Cascade Classifier object for car detection
car_cascade = cv.CascadeClassifier(cascade_src)

# Convert the image to grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Detect cars in the grayscale image
cars = car_cascade.detectMultiScale(gray, 1.1, 1)

# Draw rectangles around the detected cars
for (x, y, w, h) in cars:
    cv.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)

# Display the image with cars detected
cv.imshow('Cars Detected', img)
cv.waitKey(0)
cv.destroyAllWindows()
```
